```python
# - Load the dataset into DataFrames.
# - Handle missing values & data cleaning.
# - Analyze total sales per customer.
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['ShipDate'] = pd.to_datetime(df['ShipDate'])
```

```python
df['OrderYear'] = df['OrderDate'].dt.year
df['OrderMonth'] = df['OrderDate'].dt.month
df['OrderDay'] = df['OrderDate'].dt.day_name()
```

```python
df['ShipYear'] = df['ShipDate'].dt.year
df['ShipMonth'] = df['ShipDate'].dt.month
df['ShipDay'] = df['ShipDate'].dt.day_name()
```

```python
customer_sales = df.groupby('CustomerName')['Sales'].sum().sort_values(ascending = False)
customer_sales
```

```python
top_customers = customer_sales.head(10)
top_customers
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='viridis')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Customer Name')
plt.tight_layout()
plt.show()
```

```python
df.info()
```

```python
monthly_sales = df.groupby(pd.Grouper(key='OrderDate', freq='M'))['Sales'].sum()
monthly_sales_ma = monthly_sales.rolling(window=3).mean()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales, label='Monthly Sales', color='skyblue')
plt.plot(monthly_sales_ma, label='3-Month Moving Average', color='orange', linewidth=2)
plt.title('Monthly Sales with 3-Month Moving Average')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='OrderMonth', y='Sales', data=df)
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(monthly_sales, model='additive') 
result.plot()
plt.show()
```
